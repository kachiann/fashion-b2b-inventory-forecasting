```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#df = pd.read_csv("data/processed/test_predictions.csv", parse_dates=["week"])

feat_path = "../data/processed/test_predictions.csv"
df_test = pd.read_csv(feat_path, parse_dates=["week"])
```

```python
sample_partner = df_test["partner_id"].iloc[0]
sample_sku = df_test["sku_id"].iloc[0]
```

```python
mask = (df_test["partner_id"] == sample_partner) & (df_test["sku_id"] == sample_sku)
df_sample = df_test.loc[mask].sort_values("week")
```

```python
plt.figure(figsize=(10, 4))
plt.plot(df_sample["week"], df_sample["sales_units"], label="Actual", marker="o")
plt.plot(df_sample["week"], df_sample["pred_sales_units"], label="Predicted", marker="x")
plt.xlabel("Week")
plt.ylabel("Sales units")
plt.title(f"Actual vs predicted  {sample_partner}, {sample_sku}")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
